Sağlıklı insanların ortalama vücut ısısı için %99 güven aralığı verin.
(a) bölümünde elde edilen güven aralığı 98.6 derece değerini, doktorlar ve diğerleri tarafından belirtilen kabul edilen ortalama sıcaklığı içeriyor mu? Hangi sonuçlara varabilirsiniz? (Wackerly, Ör.8.60)

```python
mu=98.25
n=130 
ss=0.73
sigma=ss/np.sqrt(n)
```

```python
stats.norm.interval(0.99, loc=mu, scale=sigma)
```
